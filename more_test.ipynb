```python
import obspy as ob
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

station = 'RES'

path = "DATA/waveforms_list.csv"
wv_file = pd.read_csv(path)
selected_file = wv_file.loc[wv_file['sta_code'] == station]
high_file = selected_file.loc[selected_file['rf_quality'] == 1]

for i in range(len(high_file)):
    st = ob.read(f"DATA/RF/{high_file.iloc[i]['file_name']}.pkl", format='PICKLE')
    print(high_file.iloc[i]['file_name'])
    st[1].plot()
    plt.show()
```
