```python
import pandas as pd
import obspy as op
data_type = "RF"
station = "AKVQ"

path_file = "DATA/waveforms_list.csv"
df = pd.read_csv(path_file)
df = df[df["sta_code"].isin([station])].where(df["rf_quality"] == 1).dropna()
print(len(df))
for i in range(len(df)):
    st = op.read(f"DATA/{data_type}/{df.iloc[i]["file_name"]}.pkl")[1]
    st.plot()
```
